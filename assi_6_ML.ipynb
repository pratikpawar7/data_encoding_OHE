```python
'''
Q1. What is data encoding? How is it useful in data science?
'''

Data encoding is the process of converting data from one form to another. In data science, encoding is
useful for transforming categorical data into numerical data that can be used by machine learning algorithms. 
Machine learning algorithms typically require numerical input data, so encoding helps to convert categorical 
data into numerical form so that it can be used as input to the algorithms.
```

```python
'''
Q2. What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
'''

Nominal encoding, also known as label encoding, is a type of encoding used in data science 
to convert categorical data into numerical form. It assigns a unique numerical value to each
category or label in the dataset. Nominal encoding is used when the categorical variable does 
not have any inherent order or hierarchy.

Here's an example of how you can use nominal encoding in a real-world scenario. Suppose you have
a dataset containing information about different types of fruits, and you want to predict whether
a fruit is sweet or sour based on its type. The dataset contains the following columns:

Type of Fruit (categorical variable with labels such as Apple, Banana, Mango, etc.)
Sweetness (binary variable with values 1 for sweet and 0 for sour)
To use this data in a machine learning model, you need to convert the categorical variable (Type of Fruit)
into numerical form. Nominal encoding can be used to assign a unique numerical value to each type of fruit. 

For example:
                Apple: 1
                Banana: 2
                Mango: 3
                Orange: 4
                Pineapple: 5
                
Now, you can use these numerical values to train a machine learning model to predict the sweetness of a 
fruit based on its type
```

```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

df = pd.DataFrame({
       'gender' : ['male' , 'female' , 'male' , 'female' ] , 
        'age' : [23, 34, 23 , 34 ]  , 
        'tenure' : ['a', 'b' , 'c' , 'a' ] , 
        'contract_type' : ['perment' , 'temp' , 'perment' , 'temp' ] 
    })

encoded = encoder.fit_transform(df[['gender' ,'age' , 'tenure' , 'contract_type']]).toarray()

encode_df = pd.DataFrame(encoded, columns  = encoder.get_feature_names_out()  )

pd.concat([df , encode_df] , axis =1 )
```
